```python
# import the Python Image processing Library
from PIL import Image
import numpy as np
```

```python
# Create an Image object from an Image
def crop(imageFile,left,upper,right,bottom):

    img  = Image.open(imageFile) # read the image and open
    rgb = np.array(img) # take the rgb values 
    cropped_rgb = rgb[upper:bottom,left:right,:] # take the rgbs whatever we need as a sub array (3-dimensional array)
    cropped_img = Image.fromarray(cropped_rgb) # convert rgb array to image
    print(cropped_img.size)
    return cropped_img

```
